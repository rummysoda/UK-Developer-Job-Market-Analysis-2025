## 1. Importing libraries needed

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

## 2. importing data that I will be using throughout the analysis

```python
df = pd.read_csv('survey_results_public.csv')
```

```python
# Displaying the dataframe
pd.set_option('display.max_columns', None)
df.head()
```

## 3. Data Cleaning

```python
print(df.columns.tolist())
```

```python
# get rid of non employed and people not from the uk
df.Employment.unique()
```

```python
print(df.Country.unique().tolist())
```

```python
df = df[(df.Country == 'United Kingdom of Great Britain and Northern Ireland') & ((df.Employment == 'Employed') | (df.Employment == 'Independent contractor, freelancer, or self-employed'))]
```

```python
df = df[[
        'Age',
        'EdLevel',
        'WorkExp',
        'DevType',
        'Industry',
        'LanguageHaveWorkedWith',
        'DatabaseHaveWorkedWith',
        'PlatformHaveWorkedWith',
        'WebframeHaveWorkedWith',
        'ConvertedCompYearly'
]]
```

```python
df.head()
```

```python
print(df.DevType.unique().tolist())
```

```python
df.DevType = df['DevType'].replace('Other (please specify):', np.nan)
```

```python
df.DevType = df['DevType'].replace('Student', np.nan)
```

```python
df.info()
```

```python
salaries = df[(df.ConvertedCompYearly.notna())] # Using it later for salary medians and mean
```

```python
# USD to GBP
gbp = 0.77
salaries['ConvertedCompYearly'] = salaries['ConvertedCompYearly'] * gbp
```

```python
# Compare job types salaries
salaries.groupby('DevType')['ConvertedCompYearly'].median()
```

```python
# Get rid of outliers using IQR
q1 = salaries.ConvertedCompYearly.quantile(0.25)
q3 = salaries.ConvertedCompYearly.quantile(0.75)
iqr = q3 - q1
lowerFence = q1 - 1.5 * iqr
upperFence = q3 + 1.5 * iqr

salaries = salaries[(salaries.ConvertedCompYearly < upperFence) & (salaries.ConvertedCompYearly > lowerFence) ]
```

```python
# Compare job types salaries after getting rid of outliers
salaries.groupby('DevType')['ConvertedCompYearly'].median()
```

## 4. Analaysis

#### Distribution of indvidial salaries in the UK for software related development 


```python
# Distribution of indvidial salaries in the UK for software related development 
fig, (plt1,plt2) = plt.subplots(2, 1, figsize=(12,8))

plt1.hist(salaries.ConvertedCompYearly, bins = 50,color='skyblue', edgecolor='black')
plt1.axvline(salaries['ConvertedCompYearly'].median(), linestyle='--', linewidth = 2)
plt1.set_xlabel('salaries (£)')
plt1.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'£{int(x/1000)}K'))
plt1.set_xticks(range(0, 180000, 10000))
plt1.tick_params(axis='x', rotation=45)
plt1.set_ylabel('number of respondents')
plt1.set_title('Distribution of indvidial salaries in the UK for software related development')

plt2.boxplot([salaries.ConvertedCompYearly], vert=False)
plt2.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'£{int(x/1000)}K'))
plt2.set_yticks([])
plt.tight_layout()
```

#### Salary by role per job title

```python
# Takes all salary values for each role into a list
salariesByJob = salaries.groupby('DevType')['ConvertedCompYearly'].apply(list)
print(salariesByJob.index.tolist())
```

```python
# Sorting the roles by the median salary
salariesByJob = salariesByJob.reindex(salariesByJob.apply(np.median).sort_values().index)
```

```python
# Salary by role per job title
plt.figure(figsize=(9,10))
plt.boxplot(salariesByJob.values, vert = False)
plt.yticks(range(1, len(salariesByJob) + 1), salariesByJob.index)
plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'£{int(x/1000)}K'))
plt.xlabel('salaries (£)')
plt.title('Salary by role per job title')
plt.show()
```

```python
roleMedians = salaries.groupby('DevType')['ConvertedCompYearly'].median()
roleMedians = roleMedians.sort_values(ascending=False)
```

```python
'''sns.set_palette('Blues_r')
plt.figure(figsize=(12,12))
plt.barh(roleMedians.index,roleMedians.values)
plt.xlabel('salaries (£)')
plt.title("salary per job title")'''
fig, ax = plt.subplots(figsize=(12, 12))
sns.barplot(x=roleMedians.values, y=roleMedians.index, palette='dark:b', hue=roleMedians.values)
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'£{int(x/1000)}K'))
ax.set_xlabel('Median Salary (£)')
ax.set_title('Salary per Job Title')
```
